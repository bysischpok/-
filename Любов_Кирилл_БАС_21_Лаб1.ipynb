```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('database_gas.csv')
print(data.head())

print(f'Статистическая информация:\n{data.describe()}\n')
print(f'Названия столбцов:\n{data.columns}\n')
print(f'{data.info()}\n')

#Удаляем столбец с датой и временем
#Остальные столбцы числовые, пустых значений нет
data = data.drop('timestamp',axis=1)

#Гистограмма температуры
data['temperature'].plot(kind='hist', bins=20, color='green', edgecolor='black')
plt.xlabel('Температура')
plt.ylabel('Частота')
plt.show()
#Гистограмма влажности
data['humidity'].plot(kind='hist', bins=15, color='green', edgecolor='black')
plt.xlabel('Влажность')
plt.ylabel('Частота')
plt.show()
#Гистограмма CO2CosIRValue
data['CO2CosIRValue'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.xlabel('CO2CosIRValue')
plt.ylabel('Частота')
plt.show()
#Гистограмма CO2MG811Value
data['CO2MG811Value'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.xlabel('CO2MG811Value')
plt.ylabel('Частота')
plt.show()
#Гистограмма MOX1
data['MOX1'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.xlabel('MOX1')
plt.ylabel('Частота')
plt.show()
#Гистограмма MOX2
data['MOX2'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.xlabel('MOX2')
plt.ylabel('Частота')
plt.show()
#Гистограмма MOX3
data['MOX3'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.xlabel('MOX3')
plt.ylabel('Частота')
plt.show()
#Гистограмма MOX4
data['MOX4'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.xlabel('MOX4')
plt.ylabel('Частота')
plt.show()
#Гистограмма COValue
data['COValue'].plot(kind='hist', bins=30, color='green', edgecolor='black')
plt.xlabel('COValue')
plt.ylabel('Частота')
plt.show()
#Матрица корреляции для признаков
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
# Групповая визуализация
sns.lineplot(data=data, x='MOX3', y='MOX4');
plt.show()
sns.lineplot(data=data, x='MOX3', y='COValue');
plt.show()
```
